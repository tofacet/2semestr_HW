```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread, imshow
from skimage.color import rgb2gray
from skimage import exposure
```

```python
fft = lambda x: 100 * np.log(np.abs(np.fft.fftshift(np.fft.fft2(x))))

def show(pic):
    fig, axs = plt.subplots(1, 2, figsize=(10, 10))
    axs[0].imshow(pic, cmap='gray')
    axs[1].imshow(fft(pic), cmap='gray')
```

```python
pics = [imread(f'4_0{i}.jpg') for i in [1, 2, 4, 5, 6, 7]]
```

```python
[show(pic) for pic in pics]
```
